```python
import numpy as np
import matplotlib as plt #import the matrix plot library
%matplotlib inline 
```

```python
points = np.arange(-5, 5, 0.01) # arrange the array: X - Start point, Y - End point, Z - Steps taken to go through end point.
```

```python
dx, dy = np.meshgrid(points, points)
```

```python
dx
```

```python
dy
```

```python
z = (np.sin(dx) + np.sin(dy)) # use the sin method to find dx and dy.
z
```

```python
w = (np.cos(dx), np.cos(dy))
w
```

```python
plt.pyplot.imshow(z) #show an image of the sin matrix
```

```python
plt.pyplot.imshow(z) # show the title and colourbar values
plt.pyplot.colorbar()

plt.pyplot.title('sin X + sin Y') 
```

```python
# numpy where

A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
```

```python
condition = np.array([True,True,False,False])
```

```python
answer = [(A_val if cond else B_val) for A_val, B_val, cond in zip(A,B,condition)] 
'''
Using a for and if-else loop in one condition to change array values.
'''
```

```python
answer
```

```python
array2 = np.where(condition, A, B) #shorter method of the previous loop
array2
```
