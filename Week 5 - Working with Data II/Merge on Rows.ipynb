```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
df1 = DataFrame({'key': ['x','y', 'z', 'x', 'x', 'x'], 'data_set1':np.arange(6)})
df1 # create DataFrame
```

```python
df2 = DataFrame({'key': ['q', 'y', 'z'], 'data_set2': [1, 2, 3]})
df2
```

```python
pd.merge(df1, df2) #merge two data frames together
```

```python
pd.merge(df1, df2, on='key') #merging data frame base on 'key'
```

```python
pd.merge(df1, df2, on='key', how='left') #merge data frame, and plug in the 'key' values from the left value (df1)
```

```python
pd.merge(df1, df2, on='key', how='right') #same shit, but for the right value (df2)
```

```python
pd.merge(df1, df2, on='key', how='outer') #now for both values.
```

```python
df3 = DataFrame({'key': ['x', 'x', 'x', 'y', 'y', 'z'], 'data_set3': np.arange(6)}) # new data frame series
df3
```

```python
df4 = DataFrame({'key': ['y', 'y', 'x', 'x', 'z'], 'data_set4': np.arange(5)})
df4
```

```python
pd.merge(df3, df4) #its a lot of rows, because the keys consist of the same values.
```

```python
df_left = DataFrame({'key1': ['NY', 'LA', 'NY'], 'key2': ['one', 'one', 'two'], 'left_data': [10, 20, 30]})
df_left
```

```python
df_right = DataFrame({'key1': ['NY', 'NY', 'LA', 'LA'], 'key2': ['one', 'two', 'one', 'two'], 'right_data': [40, 50, 60, 70]})
df_right
```

```python
pd.merge(df_left, df_right, on=['key1', 'key2'], how='outer') # 2 data frames merged together, they have common key values.
```

```python
pd.merge(df_left, df_right, on='key1') #show the merged data frame in the 'key1' order.
```

```python
pd.merge(df_left, df_right, on='key1', suffixes=('_lefty', '_righty')) #change the suffixes of 'x', 'y' to _lefty and -righty
```
