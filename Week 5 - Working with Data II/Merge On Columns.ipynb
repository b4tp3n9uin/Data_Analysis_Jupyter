```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df_left = DataFrame({'key': ['X', 'Y', 'Z', 'X', 'Y'], 'data': range(5)})
df_left
```

```python
df_right = DataFrame({'group_data': [10, 20]}, index=['X', 'Y']) #data frame with 2 named index
df_right
```

```python
pd.merge(df_left, df_right, left_on='key', right_index=True) #make the right index from df_right data merge with df_left key.
```

```python
df_leftHR = DataFrame({'key1': ['LA', 'NY', 'NY', 'LA', 'LA'], 'key2': [10, 10, 20, 30, 20], 'data_set': np.arange(5.)})
df_leftHR
```

```python
df_rightHR = DataFrame(np.arange(10).reshape(5,2), index= [['LA', 'LA', 'NY', 'NY', 'NY'], [20, 10, 10, 10, 20]],
                      columns=['col_1', 'col_2'])
df_rightHR # create 2 colums
```

```python
pd.merge(df_leftHR, df_rightHR, left_on=['key1', 'key2'], right_index=True) #merge the index of right into 'key1' of left. and mere data frame
```
